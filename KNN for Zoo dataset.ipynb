```python
#Data Pre-Processing
data.drop(['type','animal name'], axis=1,)
```

```python
# Splitting data into X and y
X = data.drop('type', axis = 1)
y = data['type']
```

```python
data.columns
```

```python
from sklearn.model_selection import train_test_split
X=data.iloc[:,1:17]
Y=data['type']
```

```python
#traing will on 80% and testing on 20%
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=30)
```

```python
#Building KNN Model
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
     
```

```python
krange = range(1, 31)
kscores = []
for k in krange:
    knn = KNeighborsClassifier(n_neighbors=k)
    train_scores = cross_val_score(knn, x_train, y_train, cv=5)
    kscores.append(train_scores.mean())
plt.plot(krange, kscores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
     
```

```python

model = KNeighborsClassifier(n_neighbors=3)
model.fit(x_train, y_train)
print('Model Score for n_neighbors =3 ');
model.score(x_test, y_test)
```

```python
#Plotting Confusion Matrix
from sklearn.metrics import confusion_matrix
pred = model.predict(x_test)
cm = confusion_matrix(y_test, pred)
cm
```
